```python
from langchain_core.tools import tool
```

```python
@tool
def add(a : int, b : int) -> int :
    """Add 2 numbers"""
    return a + b
```

```python
result = add.invoke({"a" : 9
, "b" : 3})
print(result)
```

```python
print(add.name)
print(add.description)
print(add.args)
```
